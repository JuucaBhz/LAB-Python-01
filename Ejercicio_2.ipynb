```python
n = int(input('Defina el rango: '))
suma = 0
for i in range(1, n+1):
  suma += i
  if i % 2 == 0:
    print('Numero par:', i)
print(f'La suma de los primeros {n} números naturales es: {suma}')
```

```python
# Lista inicial de temperaturas en °C
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]

promedio = sum(temperaturas) / len(temperaturas)
mpromedio = [temp for temp in temperaturas if temp > promedio]

print("Temperaturas mayores al promedio:", mpromedio)
```

```python
suma_pares = 0
impares = 0
for i in range(1, 11):
    if i % 2 == 0:
        suma_pares += i
    else:
        impares += i

estado_final = {
    "suma_pares": suma_pares,
    "impares": impares
}

print("Estado final:", estado_final)
```

Dentro del bucle, el estado cambia cada vez que la variable i toma un nuevo valor del 1 al 10. En cada vuelta, el programa revisa si ese valor es par o impar y, según el caso, aumenta la variable correspondiente (suma_pares o impares). Así, en cada iteración las variables que acumulan las sumas se actualizan reemplazando su valor anterior por uno nuevo.

```python
productos = []
while True:
    nombre = input("Ingrese producto (o 'fin' para salir): ")
    if nombre == "fin":
        break
    precio = float(input("Precio: "))
    cantidad = int(input("Cantidad: "))

    if precio < 0 or cantidad < 0:
        print("Precio o cantidad inválida. Saltando este producto.\n")
        continue

    productos.append({"nombre": nombre, "precio": precio, "cantidad": cantidad})

# Mostrar total
total = 0
for p in productos:
    total += p["precio"] * p["cantidad"]

print("Total del inventario:", total)

if productos:
    caro = max(productos, key=lambda p: p["precio"])
    print("Producto más caro:", caro)

print('Inventario final: ', productos )
```
