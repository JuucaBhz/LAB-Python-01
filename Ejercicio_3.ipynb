##Ejercicio 3 - Aplicaci√≥n pr√°ctica (integraci√≥n + API)


```python
import requests

# Llamado a una API p√∫blica
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
usuarios = response.json()
# Procesamiento imperativo
nombres = []
for u in usuarios:
    nombres.append(u["name"])

print("Usuarios encontrados:")
for n in nombres:
    print("-", n)
```

```python
# Filtrar usuarios con correo que termina en .biz
usuarios_biz = [u for u in usuarios if u["email"].endswith(".biz")]

print("Usuarios con correo .biz:")
if usuarios_biz:
    for u in usuarios_biz:
        print("-", u["name"])
else:
    print("No se encontraron usuarios con correo .biz.")
```

### Informaci√≥n Extendida de Pok√©mon con Condicionales y Formas

```python
import requests

# Entrada de datos
nombre_pokemon = input("Ingrese el nombre de un Pok√©mon: ").lower()

# Llamado a la API
url = f"https://pokeapi.co/api/v2/pokemon/{nombre_pokemon}"
respuesta = requests.get(url)

# Verificaci√≥n del estado de la respuesta
if respuesta.status_code == 200:
    datos = respuesta.json()

    # Modelado del estado del Pok√©mon
    pokemon = {
        "nombre": datos["name"],
        "altura": datos["height"],
        "peso": datos["weight"],
        "tipos": [t["type"]["name"] for t in datos["types"]],
        "habilidades": [h["ability"]["name"] for h in datos["abilities"]]
    }

    # Salida del estado actual
    print("\nüìò Informaci√≥n del Pok√©mon")
    print(f"Nombre: {pokemon['nombre'].capitalize()}")
    print(f"Altura: {pokemon['altura']}")
    print(f"Peso: {pokemon['peso']}")
    print(f"Tipos: {', '.join(pokemon['tipos'])}")


    # Si el Pok√©mon tiene m√°s de un tipo, mostrar "Pok√©mon de tipo mixto".
    if len(pokemon['tipos']) > 1:
        print("Clasificacion de Tipo: Pok√©mon de tipo mixto")


    # Si el peso es mayor a 100, mostrar "Pok√©mon de gran tama√±o".
    if pokemon['peso'] > 100:
        print("Categoria de Peso: Pok√©mon de gran tama√±o")
    else:
        print("Categoria de Peso: Pok√©mon ligero")


    # Usar for para listar las habilidades una por una
    print("Habilidades:")
    for habilidad in pokemon['habilidades']:
        print(f"  - {habilidad.capitalize()}")

    # mostrar velocidad base
    velocidad_base = datos["stats"][0]["base_stat"]
    print(f"Velocidad Base: {velocidad_base}")

else:
    print("Pokemon no encontrado. Verifique el nombre ingresado.")
```
