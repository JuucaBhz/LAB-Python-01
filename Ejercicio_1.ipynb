```python
# Estado inicial
saldo = 500_000      # Saldo de la cuenta
deposito = 150_000   # Monto a depositar
retiro = 80_000      # Monto a retirar

# Actualización del estado
saldo = saldo + deposito   # Se deposita dinero
saldo = saldo - retiro     # Se retira dinero

# Salida del estado final
print("Saldo final de la cuenta:", saldo)
```

```python
interes = 0.02
saldo_final = saldo * interes
print(f"Saldo final con interés: ${saldo_final:.2f}")
```

##¿Qué representa el estado inicial y el estado final del programa?

*Explica como cada instruccion afecta la memoria

El estado inicial del programa es simplemente la información con la que empieza: el saldo que hay en la cuenta y los valores del depósito y el retiro antes de hacer cualquier operación. El estado final es cómo quedan esos valores después de que el programa ejecuta todas las operaciones, es decir, el saldo actualizado tras sumar el depósito, restar el retiro y aplicar el interés.

Cada instrucción afecta la memoria porque cambia el valor guardado en una variable. Cuando se hace una suma o una resta, la variable saldo guarda un nuevo valor y reemplaza el anterior. Cuando se calcula el interés, se crea o modifica otra variable que también ocupa espacio en la memoria con ese nuevo dato.

```python
# Lista inicial de temperaturas en °C
temperaturas = [22.5, 23.0, 21.8, 24.3, 25.1]

# Operaciones básicas

# Agrega una nueva temperatura al final de la lista.
print("Temperaturas:", temperaturas)
temperaturas.append(1)


# Elimina la primera temperatura registrada.
print("Temperaturas:", temperaturas)
temperaturas.pop(0)
print("Temperaturas:", temperaturas)


# Inserta una temperatura en la segunda posición.
temperaturas.insert(1, 2)
print("Temperaturas:", temperaturas)


# Muestra la lista ordenada de menor a mayor.
temperaturas.sort()
print("Temperaturas:", temperaturas)


# Calcula la diferencia entre la temperatura más alta y la más baja, y calcula el promedio de todas las temperaturas.
temp_max = max(temperaturas)
temp_min = min(temperaturas)
diferencia_max_min = temp_max - temp_min
promedio_temp = sum(temperaturas) / len(temperaturas)
print('maxima temperatura:', temp_max)
print('minima temperatura:', temp_min)
print('diferencia entre max y min:', diferencia_max_min)
print('promedio de temperaturas:', promedio_temp)


# Encuentra el índice (posición) de la temperatura 23.7.
found = False
for i in range(len(temperaturas)):
  if temperaturas[i] == 23.7:
    print(f'indice de la temperatura 23.7, en el indice: {i}')
    found = True
if not found:
  print('no se encontro la temperatura 23.7')


# Muestra las temperaturas de la 2da posicion a la 4ta posicion
temp1_4 = temperaturas[1:4]
print('temperaturas de la 2da a la 4ta posicion:', temp1_4)


# Muestra las tres primeras temperaturas.
temp1_3 = temperaturas[:3]
print('las tres primeras temperaturas:', temp1_3)
```

```python
# Lista de productos en una tienda
inventario = [
    {"nombre": "Pan", "precio": 1500, "stock": 30},
    {"nombre": "Leche", "precio": 3500, "stock": 15},
    {"nombre": "Café", "precio": 8000, "stock": 10},
]

# Muestra el nombre del primer producto.
print(inventario[0]['nombre'])

# Agrega un nuevo producto al inventario.
inventario.append({"nombre": "Galletas", "precio": 2000, "stock": 2})
print(inventario)


# Disminuye el stock de “Leche” en 2 unidades (venta).
for producto in inventario:
  if producto['nombre'] == 'Leche':
    producto['stock'] -= 2
print(inventario)


# Calcula el valor total del inventario (precio * stock por producto).
total = 0
for producto in inventario:
  total += producto['precio'] * producto['stock']

print('el valor del inventario es:', total)


# Muestra los productos cuyo stock sea menor a 10.
for producto in inventario:
  if producto['stock'] < 10:
    print(producto['nombre'])


# Aumenta el precio de todos los productos en un 10%.
for producto in inventario:
  producto['precio'] *= 1.10
print(inventario)


# Redondea los precios con round(p["precio"], 2) para evitar decimales largos.
for producto in inventario:
  producto['precio'] = round(producto['precio'], 2)
print(inventario)



# Busca un producto por nombre (ingresado por el usuario) y muestra el producto encontrado, ten en cuenta que no puede estar.
nombre = input('ingrese el nombre del producto: ')
found = False
for producto in inventario:
  if producto['nombre'] == nombre:
    print(producto)
    found = True
if not found:
  print('no se encontro el producto')


# Elimina un producto del inventario si su stock es 0.
for producto in inventario:
  if producto['stock'] == 0:
    inventario.remove(producto)
print(inventario)


# Ordena la lista por precio ascendente.
inventario.sort(key=lambda x: x['precio'])
print(inventario)
```

```python
import math

posiciones = [(0, 0), (1, 2), (3, 4)]
total = 0
for i in range(len(posiciones) - 1):
    x1, y1 = posiciones[i]
    x2, y2 = posiciones[i+1]
    segmento = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    total += segmento

print('La distancia total es:',total)
```

```python
biblioteca = {
    "libros": [
        {"titulo": "1984", "autor": "George Orwell", "prestado": False},
        {"titulo": "El Principito", "autor": "Antoine de Saint-Exupéry", "prestado": True},
    ],
    "usuarios": ["Ana", "Luis", "María"]
}

# Muestra cuántos libros están prestados y cuántos disponibles.
prestados = 0
for libro in biblioteca["libros"]:
    if libro["prestado"]:
        prestados += 1
print("Libros prestados:", prestados)
print("Libros disponibles:", len(biblioteca["libros"]) - prestados)


# Agrega un nuevo libro.
biblioteca["libros"].append({"titulo": " Metamorfosis ", "autor": " Kafka ", "prestado": False})
print(biblioteca)


# Marca como prestado el libro “1984”.
for libro in biblioteca["libros"]:
    if libro["titulo"] == "1984":
        libro["prestado"] = True
print(biblioteca)



# Muestra el estado actualizado del sistema (usa print(biblioteca)).
print('biblioteca actualizda: ', biblioteca)
```
